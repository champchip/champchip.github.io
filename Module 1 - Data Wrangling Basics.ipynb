### 1.4 Define recency function

```python
def recency(x):
    return (currentDate - x.max()).days
    #return (currentDate - x)
##currentDate = superstore['Order Date'].max() (see above)
```

```python
recency(superstore['Order Date'])
#Test function à¹à¸à¸¢à¹ à¹à¸¡à¹à¸¡à¸µ meaning à¸à¹à¸­à¸ groupby per à¸¥à¸¹à¸à¸à¹à¸²à¸à¸¶à¸à¸à¸°à¸¡à¸µ à¹à¸à¸à¸à¹à¸²à¸à¸¥à¹à¸²à¸
```

### 1.5 Recency of each customer

```python
superstore.groupby('Customer ID').agg({'Order Date':recency}).head(10)
#à¸à¸·à¸­à¹à¸­à¸² recency à¹à¸ apply à¸à¸ à¹à¸à¹à¸¥à¸° group à¸à¸¹à¸à¹à¸² max à¸à¸­à¸ dif order dat à¸à¸±à¸ max order date
#à¸à¸·à¸­à¹à¸­à¸² recency à¹à¸ apply à¸à¸ à¹à¸à¹à¸¥à¸° group à¸à¸¹à¸à¹à¸² dif à¸£à¸°à¸«à¸§à¹à¸²à¸  max order date à¸à¸­à¸à¸à¸±à¹à¸ dataset  à¸à¸±à¸ max order date à¸à¸­à¸ group => x.max()
```

### 1.6 Add frequency of each customer

```python
superstore.groupby('Customer ID').agg({'Order Date':recency, 'Order ID':'nunique'})
```

### 1.7 Add monetary of each customer

```python
superstore.groupby('Customer ID').agg({'Order Date':recency, 
                                       'Order ID':'nunique',
                                       'Sales':sum})
```

### 1.8 Rename columns

```python
superstore\
    .groupby('Customer ID')\
    .agg({'Order Date':recency,'Order ID':'nunique','Sales':sum})\
    .rename(columns={'Order Date': 'R', 'Order ID':'F', 'Sales':'M'})\
    .head()
```

### 4.3 Retrieve past 3 Order ID

```python
orderRecord = superstore[['Customer ID','Order ID', 'Order Date']]\
                .drop_duplicates()\
                .sort_values(by=['Customer ID','Order Date'])
orderRecord.head()
```

```python
orderMaxSales['MaxSales'] = 1
orderMaxSales.rename(columns={'Sales':'Order ID'},inplace=True)
orderMaxSales.head()
```

```python
orderRecord_1 = orderRecord.merge(orderMaxSales, on=['Customer ID','Order ID'], how = 'left')
#à¸à¸£à¸£à¸à¸±à¸à¸à¸ order id à¹à¸«à¸ à¹à¸à¹à¸ max sale à¸à¹à¸à¸°à¸¡à¸µà¸à¹à¸² left join à¸¡à¸² à¸à¹à¸²à¹à¸¡à¹à¹à¸à¹ max sale à¸¡à¸µ fillna =0 à¸­à¸¢à¸¹à¹à¸à¹à¸²à¸à¸¥à¹à¸²à¸
orderRecord_1.head(10)
```

```python
orderRecord_1.groupby('Customer ID')['MaxSales'].shift(-1)
```

```python
orderRecord_1 = orderRecord.merge(orderMaxSales, on=['Customer ID','Order ID'], how = 'left')
orderRecord_1['M1'] = orderRecord_1.groupby('Customer ID')['MaxSales'].shift(-1)
orderRecord_1['M2'] = orderRecord_1.groupby('Customer ID')['M1'].shift(-1)
orderRecord_1['M3'] = orderRecord_1.groupby('Customer ID')['M2'].shift(-1)
orderRecord_1 = orderRecord_1.fillna(0)
orderRecord_1['PastRecords'] = orderRecord_1['M1'] + orderRecord_1['M2'] + orderRecord_1['M3']
orderRecord_1.head(n=10)
```

### 4.4 Filter only customers that have at least 3-month records back before max

```python
orderRecord_2 = orderRecord_1.groupby('Customer ID', as_index=False).agg({'PastRecords':sum})
orderRecord_2.head()
```

```python
selectedCustomers = orderRecord_2[orderRecord_2['PastRecords'] == 3].drop(columns='PastRecords')
selectedCustomers
#à¹à¸à¹à¸¥à¸¹à¸à¸à¹à¸²à¸à¸µà¹3
```

### 4.5 Filter by (inner) joining

```python
orderRecord_3 = orderRecord_1[orderRecord_1['PastRecords'] == 1]\
                    .merge(selectedCustomers, on='Customer ID')\
                    .drop(columns=['MaxSales','PastRecords'])
orderRecord_3.head()
```

### 4.6 Melt M1-M3 to a column

```python
orderRecord_past = pd.melt(orderRecord_3, 
                           id_vars=['Customer ID','Order ID','Order Date'], 
                           value_vars=['M1','M2','M3'], 
                           var_name='Month', 
                           value_name='Value')
orderRecord_past
```

```python
orderRecord_past = pd.melt(orderRecord_3, 
                           id_vars=['Customer ID','Order ID','Order Date'], 
                           value_vars=['M1','M2','M3'], 
                           var_name='Month', 
                           value_name='Value')
orderRecord_past = orderRecord_past[orderRecord_past['Value'] > 0]\
                    .drop(columns='Value')\
                    .sort_values(by=['Customer ID','Order Date'], ascending=False)
orderRecord_past.head(n=10)

#à¸à¸¶à¸à¸à¸±à¹à¸à¸à¸µà¹à¸à¸·à¸­à¸à¸²à¸£à¸à¸³à¸«à¸à¸ target à¸¥à¸¹à¸à¸à¹à¸²à¹à¸à¹à¹à¸¥à¹à¸§à¸§à¹à¸²à¹à¸à¹à¸à¹à¸à¸£à¸à¹à¸²à¸  à¹à¸¥à¹à¸à¸à¸±à¹à¸à¸à¹à¸­à¹à¸à¸à¸·à¸­à¹à¸à¸à¸¹ profile à¸¥à¸¹à¸à¸à¹à¸²
```

### 4.7 Prepare transactions

```python
sales_profit = superstore\
                .groupby(['Customer ID','Order ID'], as_index=False)\
                .agg({'Sales':sum, 'Profit':sum})
sales_profit.head()
```

### 4.8 Join Transaction data

```python
orderRecord_past\
    .merge(sales_profit, on=['Customer ID','Order ID'], how='inner')
```

```python
orderRecord_final = orderRecord_past\
                    .merge(sales_profit, 
                           on=['Customer ID','Order ID'], 
                           how='inner')\
                    .pivot_table(index=['Customer ID'], 
                                 columns='Month', 
                                 values=['Sales','Profit'], 
                                 aggfunc=sum)
orderRecord_final
```

```python
result = None
import numpy as np
for name, group in fligthGroup :
    res = group.sort_values(by='dep_delay',ascending=False).iloc[0:10]
    #pd.concat()
    #res['carrier']
    res1 = pd.DataFrame({name:res['carrier'].str.cat(res['flight'].values.astype('str')),'ind':np.arange(0,10,1)})
    res1.set_index('ind',inplace=True)
    result = pd.concat([result,res1],axis=1)
    
result
```

```python
def get_worst_airlines(df):
    res = df.sort_values(by='dep_delay',ascending= False).iloc[0:10]
    res1 = pd.DataFrame({'name': res['carrier'].str.cat(res['flight'].values.astype('str')),'ind':np.arange(0,10,1)})
    return(res1)
```
